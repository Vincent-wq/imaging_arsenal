```python
# check sequeces
import numpy as np
import pandas as pd
import nipype.interfaces.io as nio
import matplotlib.pyplot as plt
import nibabel as nib

PD_info_path='/project/PD_INFO/'
ET_info_path='/project/ET_INFO/'
HC_info_path='/project/HC_INFO/'

pd_s1=pd.read_csv(PD_info_path+'ses-1.log', sep='\t', header=0)
pd_s2=pd.read_csv(PD_info_path+'ses-2.log', sep='\t', header=0)
et_s1=pd.read_csv(ET_info_path+'ses-1.log', sep='\t', header=0)
et_s2=pd.read_csv(ET_info_path+'ses-2.log', sep='\t', header=0)
hc_s1=pd.read_csv(HC_info_path+'ses-1.log', sep='\t', header=0)

pd_s1_uni=pd_s1.iloc[:,[2,13]]
pd_s2_uni=pd_s2.iloc[:,[2,13]]
et_s1_uni=et_s1.iloc[:,[2,13]]
et_s2_uni=et_s2.iloc[:,[2,13]]
hc_s1_uni=hc_s1.iloc[:,[2,13]]
all_seq=pd_s1_uni.append([pd_s2_uni,et_s1_uni,et_s2_uni,hc_s1_uni], ignore_index=True)

all_seq_1=all_seq.iloc[:,0]+'+'+all_seq.iloc[:,1]
all_seq_1_uni=all_seq_1.unique()
print(all_seq)
#print(all_seq_1_uni)
#all_seq_uni=all_seq.iloc[:,2].unique()
```

```python
#all_seq_1_uni
def t1w(x):
    #print(x)
    a,b=x.split('+')
    return a if 'T1 MPRAGE ADNI' in a else 0
def t2w(x):
    a,b=x.split('+')
    return a if 'T2 Axial' in a else 0
def swi(x):
    a,b=x.split('+')
    return a if ('swi' in a or 'SWI' in a) else 0
def bold(x):
    a,b=x.split('+')
    if (('BOLD' in a) or ('bold' in a)) and (('resting' in a) or ('Resting' in a)) and b=='False':
        if 'gre_field_mapping' in a:
            return a+'+fmap'
        else:
            return a+'+bold'
    else:
        return 0
def dwi(x):
    a,b=x.split('+')
    if ('diffusion' in a) or ('DIFFUSION' in a):
        if 'gre_field_mapping' in a:
            return a+'+fmap'
        else:
            return a+'+dwi'
    else:
        return 0
def showList(x):
    res=[]
    for a in x:
        if a!='0':
            print(a)
            res.append(a)
    return res
t1w_list=np.array(list(map(lambda x: t1w(x),  all_seq_1_uni)))
t2w_list=np.array(list(map(lambda x: t2w(x),  all_seq_1_uni)))
bold_list=np.array(list(map(lambda x: bold(x), all_seq_1_uni)))
dwi_list=np.array(list(map(lambda x: dwi(x),  all_seq_1_uni)))
swi_list=np.array(list(map(lambda x: swi(x),  all_seq_1_uni)))
print('T1w:'); res_t1w=showList(t1w_list)
print('T2w:'); res_t2w=showList(t2w_list)
print('bold:'); res_bold=showList(bold_list)
print('DWI:'); res_dwi=showList(dwi_list)
```
