```python
import time
from pathlib import Path
import os

out_dir = Path('/output')
res_dir = out_dir / 'template_mni_res'
if not res_dir.is_dir():
    os.mkdir(res_dir)
temp40_dir = out_dir / 'res_template40'
temp19_dir = out_dir / 'res_template19'
template_mni_file= Path('/templateflow/tpl-MNI152NLin2009cAsym/tpl-MNI152NLin2009cAsym_res-01_desc-brain_T1w.nii.gz')
brainmask_mni_file= Path('/templateflow/tpl-MNI152NLin2009cAsym/tpl-MNI152NLin2009cAsym_res-01_desc-brain_mask.nii.gz')

GROUPS=['PD','ET','NC']
ind_tmp_list = {}; mni_tmp_list = {};
for i_group in range(len(GROUPS)):
    tmp_name = GROUPS[i_group] + '_40_' + 'ind'
    tmp_file = temp40_dir /  (GROUPS[i_group] + '_ind.nii.gz')
    ind_tmp_list[tmp_name] = tmp_file
    tmp_name = GROUPS[i_group] + '_19_' + 'ind'
    tmp_file = temp19_dir /  (GROUPS[i_group] + '_ind.nii.gz')
    ind_tmp_list[tmp_name] = tmp_file
    
def ants_RegSyn(moving_file, fix_file, output_prefix, nTreads):
    from nipype.interfaces.ants import RegistrationSynQuick
    reg = RegistrationSynQuick()
    reg.inputs.fixed_image  = fix_file
    reg.inputs.moving_image = moving_file
    reg.inputs.output_prefix  = output_prefix
    # basic inputs
    reg.inputs.dimension = 3
    reg.inputs.transform_type = 's'
    # hiperparameter
    # computing
    reg.inputs.num_threads = nTreads
    # default
    #reg.inputs.use_histogram_matching = False # This is the default
    cmd_text=reg.cmdline
    print("running: ", cmd_text)
    reg.run()
    return cmd_text
```
